```python
# json to geojson
# for using solaris

import json
import os
file_list = os.listdir('')
file_list

# (length.file_list())
### target =  + file_list[i]
### save =  + file_list[i]+'r' +' '.geojson'


target = r'C:/Users/CES/FR_AP_36701090_340_FGT.json'
save = r'C:/Users/CES/FR_AP_36701090_340_FGT_r.geojson'

with open(target, 'rt', encoding='UTF8') as f:
    gj = json.load(f)
    
    replaceTarget = gj
    replaceTarget = str(replaceTarget)
    replaceTarget = replaceTarget.replace('ANN_CD', 'tree')
    
    replaceTarget = replaceTarget.replace(': 0', ': "0"')
    replaceTarget = replaceTarget.replace(', \'ANN_NM\': \'\'', '')
    
    replaceTarget = replaceTarget.replace(': 190', ': "0"')
    replaceTarget = replaceTarget.replace(', \'ANN_NM\': \'\'', '')
    
    replaceTarget = replaceTarget.replace(': 110', ': "2"')
    replaceTarget = replaceTarget.replace(', \'ANN_NM\': \'\'', '')
    
    replaceTarget = replaceTarget.replace(': 120', ': "2"')
    replaceTarget = replaceTarget.replace(', \'ANN_NM\': \'\'', '')
    
    replaceTarget = replaceTarget.replace(': 130', ': "2"')
    replaceTarget = replaceTarget.replace(', \'ANN_NM\': \'\'', '')
    
    replaceTarget = replaceTarget.replace(': 140', ': "1"')
    replaceTarget = replaceTarget.replace(', \'ANN_NM\': \'\'', '')
    
    replaceTarget = replaceTarget.replace('\'', '\"')

    replaceResult = replaceTarget
    print(replaceResult)
    
with open(save, 'w', encoding = 'cp949') as r:
    r.write(replaceResult)
```
